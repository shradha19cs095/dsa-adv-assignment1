```python
#Delete the elements in an linked list whose sum is equal to zero

class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None

def delete_zero_sum_sublists(head):
    dummy = ListNode(0)
    dummy.next = head
    current = dummy
    running_sum = 0
    sum_to_node = {}
    
    while current:
        running_sum += current.value
        
        if running_sum in sum_to_node:
            # Remove elements between sum_to_node[running_sum] and current
            node = sum_to_node[running_sum].next
            temp_sum = running_sum + node.value
            while temp_sum != running_sum:
                del sum_to_node[temp_sum]
                node = node.next
                temp_sum += node.value
            sum_to_node[running_sum].next = current.next
        else:
            sum_to_node[running_sum] = current
        
        current = current.next
    
    return dummy.next

# Example usage:
# Create a linked list: 1 -> 2 -> -3 -> 3 -> 1
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(-3)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(1)

result = delete_zero_sum_sublists(head)

# Print the result: 1 -> 2 -> 1
while result:
    print(result.value, end=" -> ")
    result = result.next
```

```python
#Reverse a linked list in groups of given size

class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None

def reverse_in_groups(head, k):
    if not head or k <= 1:
        return head
    
    dummy = ListNode(0)
    dummy.next = head
    prev_group_end = dummy
    while True:
        current_group_start = prev_group_end.next
        current = current_group_start
        i = 1
        
        # Traverse to find the end of the current group
        while current and i < k:
            current = current.next
            i += 1
        
        if not current:
            break  # Reached the end of the list
        
        next_group_start = current.next
        current.next = None  # Disconnect the current group
        
        # Reverse the current group
        current_group_start, current_group_end = reverse_linked_list(current_group_start)
        
        # Connect the reversed group back to the previous group
        prev_group_end.next = current_group_start
        current_group_end.next = next_group_start
        
        # Update pointers for the next iteration
        prev_group_end = current_group_end
    
    return dummy.next

def reverse_linked_list(head):
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev, head

# Example usage:
# Create a linked list: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8
head = ListNode(1)
current = head
for i in range(2, 9):
    current.next = ListNode(i)
    current = current.next

k = 3
result = reverse_in_groups(head, k)

# Print the result: 3 -> 2 -> 1 -> 6 -> 5 -> 4 -> 8 -> 7
while result:
    print(result.value, end=" -> ")
    result = result.next
```

```python
#Merge a linked list into another linked list at alternate positions.

class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None

def merge_alternate(head1, head2):
    if not head1:
        return head2
    if not head2:
        return head1

    current1 = head1
    current2 = head2

    while current1 and current2:
        temp1 = current1.next
        temp2 = current2.next

        current1.next = current2
        current2.next = temp1

        current1 = temp1
        current2 = temp2

    return head1

# Helper function to print a linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Example usage:
# Create two linked lists: 1 -> 2 -> 3 and A -> B -> C -> D
head1 = ListNode(1)
head1.next = ListNode(2)
head1.next.next = ListNode(3)

head2 = ListNode('A')
head2.next = ListNode('B')
head2.next.next = ListNode('C')
head2.next.next.next = ListNode('D')

merged_head = merge_alternate(head1, head2)

# Print the merged linked list: 1 -> A -> 2 -> B -> 3 -> C -> D -> None
print_linked_list(merged_head)
```

```python
#In an array, Count Pairs with given sum 

def count_pairs_with_sum(arr, target_sum):
    if not arr:
        return 0

    num_count = {}
    count = 0

    for num in arr:
        complement = target_sum - num
        if complement in num_count:
            count += num_count[complement]
        if num in num_count:
            num_count[num] += 1
        else:
            num_count[num] = 1

    return count

# Example usage:
arr = [1, 2, 3, 4, 5, 6]
target_sum = 7

result = count_pairs_with_sum(arr, target_sum)
print("Number of pairs with the sum", target_sum, "is", result)
```

```python
#Find duplicates in an array

def find_duplicates(arr):
    if not arr:
        return []

    seen = set()
    duplicates = []

    for num in arr:
        if num in seen:
            duplicates.append(num)
        else:
            seen.add(num)

    return duplicates

# Example usage:
arr = [1, 2, 3, 4, 2, 5, 6, 3, 7]
result = find_duplicates(arr)
print("Duplicates in the array:", result)
```

```python
#Find the Kth largest and Kth smallest number in an array

def find_kth_largest(arr, k):
    if not arr or k <= 0 or k > len(arr):
        return None

    # Sort the array in ascending order
    arr.sort()

    # Return the Kth largest element (Kth element from the end)
    return arr[-k]

def find_kth_smallest(arr, k):
    if not arr or k <= 0 or k > len(arr):
        return None

    # Sort the array in ascending order
    arr.sort()

    # Return the Kth smallest element
    return arr[k - 1]

# Example usage:
arr = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
k = 3

kth_largest = find_kth_largest(arr, k)
kth_smallest = find_kth_smallest(arr, k)

print(f"{k}th largest number is: {kth_largest}")
print(f"{k}th smallest number is: {kth_smallest}")
```

```python
#Move all the negative elements to one side of the array

def move_negatives_to_one_side(arr):
    if not arr:
        return arr

    left = 0
    right = len(arr) - 1

    while left <= right:
        if arr[left] < 0 and arr[right] < 0:
            # Both elements are negative, move the left pointer to the right
            left += 1
        elif arr[left] >= 0 and arr[right] < 0:
            # Left is non-negative, right is negative, swap them
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
        elif arr[left] >= 0 and arr[right] >= 0:
            # Both elements are non-negative, move the right pointer to the left
            right -= 1
        else:
            # Left is negative, right is non-negative, move both pointers
            left += 1
            right -= 1

# Example usage:
arr = [-1, 3, -2, 2, 7, -5, 9, -6]
move_negatives_to_one_side(arr)
print("Array with negative elements on one side:", arr)
```

```python
#Reverse a string using a stack data structure

def reverse_string(input_str):
    stack = []

    # Push each character of the input string onto the stack
    for char in input_str:
        stack.append(char)

    reversed_str = ''

    # Pop characters from the stack to create the reversed string
    while stack:
        reversed_str += stack.pop()

    return reversed_str

# Example usage:
input_str = "Hello, World!"
reversed_str = reverse_string(input_str)
print("Original String:", input_str)
print("Reversed String:", reversed_str)
```

```python
#Evaluate a postfix expression using stack

def evaluate_postfix(expression):
    stack = []

    operators = {'+', '-', '*', '/'}

    for token in expression.split():
        if token not in operators:
            # If it's an operand (number), push it onto the stack
            stack.append(int(token))
        else:
            # If it's an operator, pop the top two operands from the stack,
            # perform the operation, and push the result back onto the stack
            operand2 = stack.pop()
            operand1 = stack.pop()
            if token == '+':
                result = operand1 + operand2
            elif token == '-':
                result = operand1 - operand2
            elif token == '*':
                result = operand1 * operand2
            elif token == '/':
                result = operand1 / operand2
            stack.append(result)

    # The final result should be the only element in the stack
    return stack[0]

# Example usage:
postfix_expression = "3 4 + 2 * 7 /"
result = evaluate_postfix(postfix_expression)
print("Result of postfix expression:", result)
```

```python
#Implement a queue using the stack data structure

class QueueUsingStacks:
    def __init__(self):
        self.stack1 = []  # For enqueue operation
        self.stack2 = []  # For dequeue operation

    def enqueue(self, item):
        self.stack1.append(item)

    def dequeue(self):
        if not self.stack2:
            # If stack2 is empty, pop all elements from stack1 and push them to stack2
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        
        if self.stack2:
            # Pop the front element from stack2 (which corresponds to the front of the queue)
            return self.stack2.pop()
        else:
            # If both stacks are empty, the queue is empty
            return None

    def is_empty(self):
        return not (self.stack1 or self.stack2)

    def size(self):
        return len(self.stack1) + len(self.stack2)

# Example usage:
queue = QueueUsingStacks()

queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print("Dequeue:", queue.dequeue())  # Dequeue: 1
print("Dequeue:", queue.dequeue())  # Dequeue: 2

queue.enqueue(4)
print("Size of the queue:", queue.size())  # Size of the queue: 2
print("Is the queue empty?", queue.is_empty())  # Is the queue empty? False

print("Dequeue:", queue.dequeue())  # Dequeue: 3
print("Dequeue:", queue.dequeue())  # Dequeue: 4
print("Is the queue empty?", queue.is_empty())  # Is the queue empty? True
```
